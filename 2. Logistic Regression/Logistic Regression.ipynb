## Training the model

```python
# Prepare data for modeling
X = df.drop('Bankrupt', axis=1).values
y = df['Bankrupt'].values
```

```python
X[0]
```

```python
def split_data(X, y, test_size=0.2, random_state=42):
    np.random.seed(random_state)
    n = X.shape[0]
    indices = np.arange(n)
    np.random.shuffle(indices)
    
    test_size = int(test_size * n)
    test_indices = indices[:test_size]
    train_indices = indices[test_size:]
    
    X_train, X_test = X[train_indices], X[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = split_data(X, y)
```

```python
# Compute training statistics
train_mean = X_train.mean(axis=0)
train_std = X_train.std(axis=0)
std_safe = np.where(train_std == 0, 1, train_std)

# Standardize using training stats
X_train = (X_train - train_mean) / std_safe
X_test = (X_test - train_mean) / std_safe

# Add bias column (intercept term)
X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
w = np.zeros(X_train.shape[1])
alpha = 0.1
epochs = 1000
batch_size = 32
```

```python
for epoch in range(1, epochs + 1):
    # Shuffle each epoch
    indices = np.arange(X_train.shape[0])
    np.random.shuffle(indices)
    X_train = X_train[indices]
    y_train = y_train[indices]
    m_train = X_train.shape[0]

    for start in range(0, m_train, batch_size):
        end = start + batch_size
        X_batch = X_train[start:end]
        y_batch = y_train[start:end]

        m_batch = X_batch.shape[0]
        z = X_batch @ w
        y_pred = sigmoid(z)

        dw = (1 / m_batch) * X_batch.T @ (y_pred - y_batch)
        w -= alpha * dw

    if epoch % 100 == 0:
        z_train = X_train @ w
        y_train_pred = sigmoid(z_train)
        train_loss = - (1 / m_train) * np.sum(
            y_train * np.log(y_train_pred + 1e-15) +
            (1 - y_train) * np.log(1 - y_train_pred + 1e-15)
        )
        print(f'Epoch {epoch}, Loss: {train_loss:.4f}')
```
